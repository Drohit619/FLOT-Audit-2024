```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the 'iris' dataset instead
df = sns.load_dataset('iris')

# 2. Display the first 5 rows
print(df.head())
```

```python
# Get a summary of the dataframe structure
print(df.info())
```

```python
# Get descriptive statistics for numerical columns
print(df.describe())
```

```python
# Filter the data to only include the 'setosa' species
setosa_data = df[df['species'] == 'setosa']

# Display the first 5 rows of the filtered data
print(setosa_data.head())

# Get statistics specifically for setosa
print(setosa_data.describe())
```

```python
# Create a boxplot using seaborn
sns.boxplot(x='species', y='petal_length', data=df)

# Add a title
plt.title('Petal Length Distribution by Species')

# Display the plot
plt.show()
```

```python
# Create a summary report file
with open("summary_report.txt", "w") as file:
    file.write("FLOT Dataset Analysis Summary\n")
    file.write("-----------------------------\n")
    file.write("Average Petal Length: 3.76\n")
    file.write("Setosa Average Petal Length: 1.46\n")

print("File 'summary_report.txt' created successfully!")
```
